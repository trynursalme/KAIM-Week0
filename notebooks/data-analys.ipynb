```python
import pandas as pd
```

## Load the data

```python
data_benin= pd.read_csv("../data/benin-malanville.csv")
data_sier = pd.read_csv("../data/sierraleone-bumbuna.csv")
data_togo = pd.read_csv("../data/togo-dapaong_qc.csv")
```

## Add new column " Country "to each dataset

```python
# Add a "country" column
data_benin['Country'] = 'Benin'
data_sier['Country'] = 'Sierra Leone'
data_togo['Country'] = 'Togo'
```

```python
# Merge datasets
merged_data = pd.concat([data_benin,data_sier,data_togo], ignore_index=True)
```

```python
# Convert Timestamp to datetime format
merged_data['Timestamp'] = pd.to_datetime(merged_data['Timestamp'])

# Set Timestamp as the index
merged_data.set_index('Timestamp', inplace=True)
```

```python
merged_data.head()
```

```python
missing_values = merged_data.isnull().sum()
print(missing_values)
```

```python
print(merged_data['Comments'].isnull().sum())
```

```python
merged_data = merged_data.drop(columns=['Comments'])
```

```python
missing_values = merged_data.isnull().sum()
print(missing_values)
```

```python
outliers = merged_data[(merged_data['GHI'] < 0) | (merged_data['DNI'] < 0)]
print(outliers)
```

```python
## Checking the outliers by the country
outliers_by_country = outliers['Country'].value_counts()
print(outliers_by_country)
```

```python
# For the outlier will be indicative of missing value data
# Input the negative values with the median of GHI, DNI, and DHI
median_ghi = merged_data[merged_data['GHI'] >= 0]['GHI'].median()
median_dni = merged_data[merged_data['DNI'] >= 0]['DNI'].median()
median_dhi = merged_data[merged_data['DHI'] >= 0]['DHI'].median()

merged_data.loc[merged_data['GHI'] < 0, 'GHI'] = median_ghi
merged_data.loc[merged_data['DNI'] < 0, 'DNI'] = median_dni
merged_data.loc[merged_data['DHI'] < 0, 'DHI'] = median_dhi
```

```python
summary_stats = merged_data.describe()
print(summary_stats)
```

```python
outliers = merged_data[(merged_data['GHI'] < 0) | (merged_data['DNI'] < 0)]
print(outliers)
```

```python
import matplotlib.pyplot as plt
```

```python
# Resample data to daily averages for better visualization
# slect only numeric columns
numeric_data = merged_data.select_dtypes(include=['float64', 'int64'])
daily_data = numeric_data.resample('D').mean()
```

```python
# Plot GHI, DNI, DHI, and Tamb over time
plt.figure(figsize=(14, 8))

# Subplot 1: GHI over time
plt.subplot(2, 2, 1)
plt.plot(daily_data.index, daily_data['GHI'], label='GHI', color='orange')
plt.title('Global Horizontal Irradiance (GHI) Over Time')
plt.xlabel('Date')
plt.ylabel('GHI (W/m)')
plt.grid()

# Subplot 2: DNI over time
plt.subplot(2, 2, 2)
plt.plot(daily_data.index, daily_data['DNI'], label='DNI', color='red')
plt.title('Direct Normal Irradiance (DNI) Over Time')
plt.xlabel('Date')
plt.ylabel('DNI (W/m)')
plt.grid()

# Subplot 3: DHI over time
plt.subplot(2, 2, 3)
plt.plot(daily_data.index, daily_data['DHI'], label='DHI', color='blue')
plt.title('Diffuse Horizontal Irradiance (DHI) Over Time')
plt.xlabel('Date')
plt.ylabel('DHI (W/m)')
plt.grid()

# Subplot 4: Tamb over time
plt.subplot(2, 2, 4)
plt.plot(daily_data.index, daily_data['Tamb'], label='Tamb', color='green')
plt.title('Ambient Temperature (Tamb) Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
# Separate data into cleaned and not cleaned subsets
cleaned_data = merged_data[merged_data['Cleaning'] == 1]
not_cleaned_data = merged_data[merged_data['Cleaning'] == 0]

# plot ModA and ModB for cleaned vs not cleaned data
plt.figure(figsize=(14,6))

# Subplot 1: ModA comparison
plt.subplot(1, 2, 1)
plt.plot(cleaned_data.index, cleaned_data['ModA'], label='Cleaned', color='blue', alpha=0.7)
plt.plot(not_cleaned_data.index, not_cleaned_data['ModA'], label='Not Cleaned', color='red', alpha=0.7)
plt.title('Impact of Cleaning on ModA')
plt.xlabel('Date')
plt.ylabel('ModA (W/m)')
plt.legend()
plt.grid()

# Subplot 2: ModB comparison
plt.subplot(1, 2, 2)
plt.plot(cleaned_data.index, cleaned_data['ModB'], label='Cleaned', color='blue', alpha=0.7)
plt.plot(not_cleaned_data.index, not_cleaned_data['ModB'], label='Not Cleaned', color='red', alpha=0.7)
plt.title('Impact of Cleaning on ModB')
plt.xlabel('Date')
plt.ylabel('ModB (W/m)')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

## Correlation Analysis

```python
import seaborn as sns

# Selecting relevant columns for correlation analysis
correlation_columns = ['GHI', 'DNI', 'DHI', 'TModA','TModB']
# Calculate the correlation matrix
correlation_matrix = merged_data[correlation_columns].corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Solar Radiation and Temperature')
plt.show()
```

```python
# Create pair plots for solar radiation and temperature
sns.pairplot(merged_data[correlation_columns])
plt.suptitle('Pairwise Relationships Between Solar Radiation and Temperature', y=1.02)
plt.show()
```

## Wind Analysis

```python
import numpy as np
```

```python
# Convert wind direction (WD) to radians which polar points require angles in radians so convert the WD (degrees) to radians
merged_data['WD_radians'] = np.deg2rad(merged_data['WD'])
```

```python
# Drop rows with missing or invalid wind data
clean_wind_data = merged_data.dropna(subset=['WS', 'WD'])
```

```python
import matplotlib.pyplot as plt

# Set up the polar plot
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)

# Scatter plot of wind speed and direction
ax.scatter(clean_wind_data['WD_radians'], clean_wind_data['WS'], c=clean_wind_data['WS'], cmap='viridis', alpha=0.7)

# Add labels and title
ax.set_theta_zero_location('N')  # Set 0 (North) at the top
ax.set_theta_direction(-1)       # Clockwise direction
ax.set_title('Wind Speed and Direction Distribution', va='bottom')

# Add a color bar to indicate wind speed
cbar = plt.colorbar(ax.collections[0], orientation='vertical', pad=0.1)
cbar.set_label('Wind Speed (m/s)', rotation=270, labelpad=15)

plt.show()
```

## Temperature Analysis

```python
# Drop rows with the missing values in relevant columns for the T analysis
t_clean_data = merged_data[['RH' ,'Tamb', 'GHI', 'DNI', 'DHI']].dropna()
print(t_clean_data)
```

```python
# We have to calculate correlation matrix
correlation_cal_t = t_clean_data.corr()

# Display the correlations with RH
rh_correlations = correlation_cal_t.loc['RH']
print(rh_correlations)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create scatter plots
plt.figure(figsize=(14, 6))

# Subplot 1: RH vs Tamb
plt.subplot(1, 3, 1)
sns.scatterplot(x=t_clean_data['RH'], y=t_clean_data['Tamb'], alpha=0.5)
plt.title('RH vs Ambient Temperature (Tamb)')
plt.xlabel('Relative Humidity (%)')
plt.ylabel('Temperature (C)')

# Subplot 2: RH vs GHI
plt.subplot(1, 3, 2)
sns.scatterplot(x=t_clean_data['RH'], y=t_clean_data['GHI'], alpha=0.5)
plt.title('RH vs Global Horizontal Irradiance (GHI)')
plt.xlabel('Relative Humidity (%)')
plt.ylabel('GHI (W/m)')

# Subplot 3: RH vs DNI
plt.subplot(1, 3, 3)
sns.scatterplot(x=t_clean_data['RH'], y=t_clean_data['DNI'], alpha=0.5)
plt.title('RH vs Direct Normal Irradiance (DNI)')
plt.xlabel('Relative Humidity (%)')
plt.ylabel('DNI (W/m)')

plt.tight_layout()
plt.show()
```
